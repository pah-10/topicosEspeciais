Com interação do usuário para trazer os modelos com API https://parallelum.com.br/fipe/api/v1/carros/marcas/{Número do Codigo da marca}/modelos

Exemplo: https://parallelum.com.br/fipe/api/v1/carros/marcas/21/modelos

```python
import json
import requests

#Função que verifica valores inseridos
def info_valor():
    while True:
        try:
            return int(input("Entre com o valor: "))
        except ValueError:
            print("O valor informado não é um número")

print("Informe o código da marca")
cod = info_valor()
#print(cod)

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{cod}/modelos'
#print(url)

navegador = {
    "User-Agent" : "Chrome"
}

data = requests.get(url=url, headers=navegador)

#print(data.json())

#printa os dados por partes em linha a linha
for marca in data.json()["modelos"]:
    print("\nO modelo é {} \nO código é {}".format(marca["nome"],marca["codigo"]))
```

Consumindo ano do modelo de uma marca, API: https://parallelum.com.br/fipe/api/v1/carros/marcas/{numero do codigo da marca}/modelos/{número do codigo do modelo}/anos

Exemplo: https://parallelum.com.br/fipe/api/v1/carros/marcas/21/modelos/6885/anos

```python
import json
import requests

#Função que verifica valores inseridos
def info_valor():
    while True:
        try:
            return int(input("Entre com o valor: "))
        except ValueError:
            print("O valor informado não é um número")

print("Informe o código da marca")
codMarca = info_valor()
print(codMarca)

print("Informe o código do modelo")
codModel = info_valor()
print(codModel)

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{codMarca}/modelos/{codModel}/anos'
#print(url)

navegador = {
    "User-Agent" : "Chrome"
}

data = requests.get(url=url, headers=navegador)

#print(data.json())

#printa os dados por partes em linha a linha
for marca in data.json():
    print("Esse modelo é do ano {}".format(marca["codigo"]))
```

Traz valor, API: https://parallelum.com.br/fipe/api/v1/carros/marcas/{numero do codigo da marca}/modelos/{número do codigo do modelo}/anos/{codigo ano}

Exemplo: https://parallelum.com.br/fipe/api/v1/carros/marcas/21/modelos/6885/anos/2017-1

```python
import json
import requests

def info_valor():
    while True:
        try:
            return int(input("Entre com o valor: "))
        except ValueError:
            print("O valor informado não é um número")

print("Informe o código da marca")
codMarca = info_valor()
print(codMarca)

print("Informe o código do modelo")
codModel = info_valor()
print(codModel)

codAno = input("Informe o código do ano do modelo")
print(codAno)

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{codMarca}/modelos/{codModel}/anos/{codAno}'
#print(url)

navegador = {
    "User-Agent" : "Chrome"
}

data = requests.get(url=url, headers=navegador)


print("O carro custa {}".format(data.json()["Valor"]))
```

Codigo que traz todas as infromações a partir do código da marca informado pelo usuário

```python
import os
import json
import requests

def limpar_tela():
    os.system('cls')

def info_valor():
    while True:
        try:
            return int(input("Entre com o valor: "))
        except ValueError:
            print("O valor informado não é um número")

navegador = {
    "User-Agent" : "Chrome"
}

print("Informe o código da marca: ")
codMarca = info_valor()

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{codMarca}/modelos'
 
data = requests.get(url=url, headers=navegador)

for marca in data.json()["modelos"]:
    print("\nO modelo é {} tem o código {}".format(marca["nome"],marca["codigo"]))


print("Escolha um dos códigos do modelo para prosseguir: ")
codModel = info_valor()

limpar_tela()
print(codModel)

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{codMarca}/modelos/{codModel}/anos'

data = requests.get(url=url, headers=navegador)

for marca in data.json():
    print("Esse modelo é do ano {}".format(marca["codigo"]))

codAno = input("Informe o código do ano do modelo")
limpar_tela()
print(codAno)

url = f'https://parallelum.com.br/fipe/api/v1/carros/marcas/{codMarca}/modelos/{codModel}/anos/{codAno}'

data = requests.get(url=url, headers=navegador)

print("O carro desse ano custa {}".format(data.json()["Valor"]))
```
